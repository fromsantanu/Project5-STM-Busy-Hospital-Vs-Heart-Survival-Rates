# Correlation between number of heart surgeries conducted and survival rate of patients.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Hypothetical data
data = {
    'Hospital': ['H1', 'H2', 'H3', 'H4', 'H5', 'H6', 'H7', 'H9', 'H8', 'H10'],
    'Surgeries_Performed': [200, 350, 150, 400, 500, 300, 250, 600, 700, 450],
    'Survival_Rate': [0.88, 0.85, 0.76, 0.99, 0.98, 0.97, 0.93, 0.92, 0.91, 0.90]
}

# Create a DataFrame
df = pd.DataFrame(data)

print ("Sample Data Considered for the exercise")
print(df.head())
print()
# Calculate the correlation coefficient
correlation_coefficient = np.corrcoef(df['Surgeries_Performed'], df['Survival_Rate'])[0, 1]
print("Correlation Coefficient : ",correlation_coefficient)
print("Data showing a weak positive correlation ..")

# Scatter plot with a line of best fit
plt.figure(figsize=(10, 6))
plt.scatter(df['Surgeries_Performed'], df['Survival_Rate'], c='blue', alpha=0.5, label='Data points')
print()
# Fit a line
slope, intercept = np.polyfit(df['Surgeries_Performed'], df['Survival_Rate'], 1)
plt.plot(df['Surgeries_Performed'], slope * df['Surgeries_Performed'] + intercept, color='red', label='Line of best fit')

plt.title('Relationship between Hospital Busyness and Heart Surgery Survival Rates')
plt.xlabel('Number of Heart Surgeries Performed')
plt.ylabel('Survival Rate')
plt.grid(True)
plt.legend()

# Adding labels for each point
for i, row in df.iterrows():
    plt.text(row['Surgeries_Performed'], row['Survival_Rate'], row['Hospital'])

plt.show()
```
